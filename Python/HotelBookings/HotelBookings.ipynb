```python
# Change 'reservation_status_date' column from type object to type datetime

df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='mixed', dayfirst=True)
```

```python
# get all unique values in each column

for col in df.describe(include = 'object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
# get the number of null values in each column

df.isnull().sum()
```

```python
# remove columns

df.drop(['company', 'agent'], axis=1, inplace=True)
```

```python
# remove rows

df.dropna(inplace=True)
```

```python
df.describe()
```

```python
# remove outlier

df = df[df['adr']<5000]
```

## Data Analysis and Visualizations

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation Status Count')
plt.bar(['Not cancelled', 'Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
```

```python
plt.figure(figsize = (5,7))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels = ax1.get_legend_handles_labels()
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel('No. of reservations')
plt.legend(['Not cancelled', 'Cancelled'])
```

```python
# resort hotel cancellation percentage

resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# city hotel cancellation percentage

city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel_daily = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel_daily = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel_daily.index, resort_hotel_daily['adr'], label = 'Resort Hotel')
plt.plot(city_hotel_daily.index, city_hotel_daily['adr'], label = 'City Hotel')
plt.legend(fontsize=20)
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df)
plt.legend(['Not cancelled', 'Cancelled'])
plt.xlabel('Month')
plt.ylabel('No. of reservations')
plt.title('Reservation Status Per Month', size = 20)
```

```python
plt.figure(figsize = (15,8))
plt.title('Average Daily Rate Per Month', fontsize=20)
plt.bar('month', 'adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_countries = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 Countries With Reservation Cancellations')
plt.pie(top_10_countries, autopct = '%.2f', labels = top_10_countries.index)
```

```python
cancelled_data['market_segment'].value_counts()
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

non_cancelled_data = df[df['is_canceled'] == 0]
non_cancelled_df_adr = non_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
non_cancelled_df_adr.reset_index(inplace=True)
non_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(non_cancelled_df_adr['reservation_status_date'], non_cancelled_df_adr['adr'], label = 'Non cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'Cancelled')
plt.legend()
```
